# train.csv

```python
df = pd.read_csv('../data/signate/train.tsv', sep='\t')
```

```python
df.head()
```

```python
df['datetime'].unique()
```

```python
len(df['datetime'].unique())
```

```python
df['park'].unique()
```

```python
len(df) == len(df['datetime'].unique()) * len(df['park'].unique())
```

There are 731 days and 8 parks in the training dataset. Total number of records are 731 * 8 (no missing data)

```python
df_visitors_by_day = df.groupby('datetime').agg({'visitors':['sum']})
```

```python
df_visitors_by_day.head()
```

```python
df_visitors_by_day.columns = ['visitors']
```

```python
df_visitors_by_day
```

```python
df_visitors_by_day.head(108).plot()
```

```python
df_visitors_by_day.plot()
```

```python
df_visitors_by_day[df_visitors_by_day['visitors'] > 150000]
```

Visitor number shows a highly periodical pattern. Clearly, there are significantly more visitors on some specific holidays such as new year, golden week, and Sep 20.

```python
df_visitors_by_park = df.groupby('park').agg({'visitors': 'sum'})
```

```python
df_visitors_by_park.plot(kind='bar')
```

There is a big difference of visitor number between different parks.

```python
figs, axes = plt.subplots(nrows=4, ncols=2, figsize=(20, 30))
park_names = list(df['park'].unique())

for i in range(8):
    df[df['park'] == park_names[i]].plot(ax=axes[i % 4, i // 4], x='datetime')
    axes[i % 4, i // 4].set_title(park_names[i])
plt.show()
```

Different parks have different peak dates. In general, there will be a quite large difference between crowded times and less crowded times.
